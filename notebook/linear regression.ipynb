```python
# importing librarys
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# read the dataset
df=pd.read_csv('dataset/height-weight.csv')
```

```python
df.head()
```

```python
# check for null in dataframe
df.isnull().sum()
```

```python
df.shape
```

```python
# creating dependent and independent variables.
X = df[['Weight']]
y = df['Height']
```

```python
# splitting to train test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
## Standardize the dataset
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# initialisation of regression 
regressor=LinearRegression()
```

```python
# fitting regression with training dataset
regressor.fit(X_train,y_train)
```

```python
# predicting with testing data
y_pred=regressor.predict(X_test)
```

```python
# analysing the y test and y pred, if it follows a linear pattern than its a good result
plt.scatter(y_test, y_pred)
```

```python
# printing accuracy
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
print(score)
```
